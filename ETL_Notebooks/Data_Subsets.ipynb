# Create five subsets of data to test different features #

```python
# Import dependencies

import pandas as pd
```

```python
# Create initial dataframe from cleaned data

df = pd.read_csv('../Data_Cleaned/Joined_df_cleaned.csv')
df
```

```python
df['Patient Group'].value_counts()
```

```python
# Create users df - random sample of 10 rows
users = df.sample(n=10)
users
```

```python
# Remove users from df
df = pd.merge(df, users, indicator=True, how='outer').query('_merge=="left_only"').drop('_merge', axis=1)
df = df.drop('Unnamed: 0', axis=1)
df.to_csv('../Data_Cleaned/train_test_set.csv', index = False)
df
```

```python
#  Save users df to csv
users = users.drop('Unnamed: 0', axis=1)
users.to_csv('../Data_Cleaned/User_Samples/users.csv', index = False)
users
```

## 1. Clinical information only

```python
clinical_df = df[['Patient Group', 'Age at Collection', 'BMI (kg/m2)', 'Gender_F', 'Gender_M']]
clinical_df.to_csv('../Data_Cleaned/clinical_only.csv', index= False)
clinical_df
```

## 2. JAC protein levels

```python
clinic_JAC_df = df.loc[:, ~df.columns.str.contains('AAL|NPL|Ratio|Protein|Unnamed')]
clinic_JAC_df.to_csv('../Data_Cleaned/clinic_proteins_JAC.csv', index= False)
clinic_JAC_df
```

## 3. All protein levels

```python
clinic_proteins_df = df.loc[:, ~df.columns.str.contains('Ratio|Protein|Unnamed')]
clinic_proteins_df.to_csv('../Data_Cleaned/clinic_proteins.csv', index= False)
clinic_proteins_df
```

## 4. All protein ratios

```python
clinic_ratios_df = df.loc[:, df.columns.str.contains('Patient|Ratio|Age|BMI|Gender')]
clinic_ratios_df.to_csv('../Data_Cleaned/clinic_ratios.csv', index= False)
clinic_ratios_df
```

## 5. JAC protein ratios

```python
clinic_ratios_JAC = clinic_ratios_df.loc[:, ~clinic_ratios_df.columns.str.contains('NPL|AAL')]
clinic_ratios_JAC.to_csv('../Data_Cleaned/clinic_ratios_JAC.csv', index= False)
clinic_ratios_JAC
```

```python
#  Create function to create and save data subsets
def create_subsets(input_df, output):
    clinical_df = input_df[['Patient Group', 'Age at Collection', 'BMI (kg/m2)', 'Gender_F', 'Gender_M']]
    clinical_df.to_csv(f'../Data_Cleaned/{output}_clinical_only.csv', index= False)
    clinic_JAC_df = input_df.loc[:, ~input_df.columns.str.contains('AAL|NPL|Ratio|Protein|Unnamed')]
    clinic_JAC_df.to_csv(f'../Data_Cleaned/{output}_proteins_JAC.csv', index= False)
    clinic_proteins_df = input_df.loc[:, ~input_df.columns.str.contains('Ratio|Protein|Unnamed')]
    clinic_proteins_df.to_csv(f'../Data_Cleaned/{output}_proteins.csv', index= False)
    clinic_ratios_df = input_df.loc[:, input_df.columns.str.contains('Patient|Ratio|Age|BMI|Gender')]
    clinic_ratios_df.to_csv(f'../Data_Cleaned/{output}_ratios.csv', index= False)
    clinic_ratios_JAC = clinic_ratios_df.loc[:, ~clinic_ratios_df.columns.str.contains('NPL|AAL')]
    clinic_ratios_JAC.to_csv(f'../Data_Cleaned/{output}_ratios_JAC.csv', index= False)
    
```

```python
# Create data subsets for users df
create_subsets(users, 'User_Samples/user')
```
